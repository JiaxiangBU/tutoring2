```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.random.binomial(1,0.5,100).reshape(10,-1))
df2 = pd.DataFrame(np.random.binomial(1,0.5,100).reshape(10,-1))
df3 = pd.DataFrame(np.random.binomial(1,0.5,100).reshape(10,-1))
```

```python
import os
```

```python
os.listdir('data')[0:3]
```

```python
df_list = []
for i in os.listdir('data')[0:3]:
    df_path = os.path.join('data',i)
    d = pd.read_csv(df_path)
    d['groups'] = i
    df_list.append(d)
```

```python
df = pd.concat(df_list)
```

```python
df.head()
```

```python
for idx, i in df.groupby('groups'):
    file_name = 'output' + str(idx) + '.csv'
    i.to_csv(os.path.join('data',file_name))
```
